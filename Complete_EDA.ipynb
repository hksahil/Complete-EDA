```python
## ML Methods to handle outliers
```

```python
from sklearn.cluster import DBSCAN
df['cog$'].fillna(df['cog$'].mean(),inplace=True)
```

```python
df['gross income'].fillna(df['gross income'].median(),inplace=True)
df['Total'].fillna(df['Total'].median(),inplace=True)
```

```python
df['outliers']=DBSCAN(eps=1.0, min_samples=2).fit_predict(df[['gross income']])
outllier_col=df[df['outliers']==-1]
```

```python
outllier_col['gross income']
```

```python
df.dtypes
```

```python
from sklearn.ensemble import IsolationForest
df['outliers_total']=IsolationForest(contamination=0.0001).fit_predict(df[['Total']])
outllier_col=df[df['outliers_total']==-1]
```
